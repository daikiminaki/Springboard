# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Functions

```python
#Data to ECDF
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
#Bootstrap Sampling
def bootstrap_replicate_1d(data, func):
    return func(np.random.choice(data, size=len(data)))

#Repeat bootstrap x amount of times based on size
def draw_bs_reps(data, func, size=1):
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)
        
    return bs_replicates
```

```python
#Permutation Sampling
def diff_of_means(data1, data2):
    return np.mean(data1) - np.mean(data2)

def permutation_sample(data1, data2):
    """Generate a permutation sample from two data sets."""

    # Concatenate the data sets: data
    data = np.concatenate((data1, data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1, perm_sample_2

def draw_perm_reps(data1, data2, func, size=1):
    
    perm_replicates = np.empty(size)
    
    for i in range(size):
        perm_sample_1, perm_sample_2 = permutation_sample(data1, data2)
        
        perm_replicates[i] = func(perm_sample_1, perm_sample_2)
        
    return perm_replicates
```

```python
#Pearson R
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x,y)

    # Return entry [0,1]
    return corr_mat[0,1]
```

# Hypothesis Testing: Permutation Testing
<strong>Null Hypothesis</strong>: Small and Large hospitals have no difference in excess readmission rates.<br>
<strong>Alternative Hypothesis</strong>: Small and Large hospitals have a difference in excess readmission rates.

### Assume size does not impact excess readmission rate.
This can also be interpretted as the (mean of small hospital) - (mean of large hospital) = 0

```python
#Calculate the true difference of means of small and large hospitals
true_difference = np.mean(small_hospital['Excess Readmission Ratio']) - np.mean(large_hospital['Excess Readmission Ratio'])
print('True Difference of Means: ', true_difference)
```

```python
#Get Difference of Means of Permutation Replicates
diff_perm_reps = draw_perm_reps(small_hospital['Excess Readmission Ratio'].dropna(), large_hospital['Excess Readmission Ratio'].dropna(), diff_of_means, 5000)
```

```python
_ = plt.figure(figsize=(14,7))
_ = plt.hist(diff_perm_reps, bins=100)
_ = plt.title('Difference of Means of Small and Large Hospitals')
_ = plt.axvline(np.mean(diff_perm_reps), color='red')
_ = plt.axvline(true_difference, color='blue')
```

### Mean, Standard Deviation, & Confidence Intervals of Difference of Means

```python
print('Mean: ', np.mean(diff_perm_reps))
print('STD: ', np.std(diff_perm_reps))
print('99% Confidence Interval: ', np.percentile(diff_perm_reps, [0.5,99.5]))
```

### z-score

```python
print('z-score: ', (true_difference - np.mean(diff_perm_reps)) / np.std(diff_perm_reps))
```

### P-Value

```python
print('P-Value: ', np.sum(diff_perm_reps >= true_difference) / len(diff_perm_reps))
```
